```python
##importando bibiliotecas
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
##teste
print ("teste de print")
```

```python
#criando base de Ordens de Serviço
# -------------------------
# Parâmetros
# -------------------------
N = 1000
np.random.seed(42)
# -------------------------
# Dados básicos
# -------------------------
OrderID = np.arange(1, N+1)
EquipmentID = [f"COLH-{str(i).zfill(3)}" for i in np.random.randint(1, 50, N)]
EquipmentType = ["Colhedora"] * N
OrderType = np.random.choice(["Corretiva", "Preventiva"], N, p=[0.65, 0.35])

Priority = np.random.randint(1, 6, N)
```

```python
# Datas
start_date = datetime(2024, 1, 1)
CreationDate = [start_date + timedelta(days=int(np.random.randint(0, 365))) for _ in range(N)]
FinishDate = [date + timedelta(days=int(np.random.randint(1, 30))) for date in CreationDate]
```

```python
# Status
Status = np.random.choice(["Aberta", "Em execução", "Concluída"], N, p=[0.10, 0.20, 0.70])
```

```python
# Horas de parada (maiores para corretivas)
DowntimeHours = [
    round(abs(np.random.normal(8, 3)) if t == "Corretiva" else abs(np.random.normal(3, 1)), 2)
    for t in OrderType
]
```

```python
# Custos
EstimatedCost = np.round(np.random.normal(5000, 1500, N), 2)
ActualCost = EstimatedCost + np.round(np.random.normal(300, 800, N), 2)
```

```python
# Equipes
MaintenanceGroup = np.random.choice(["Mecânica", "Elétrica", "Hidráulica"], N)
```

```python
# Descrição
Description = [
    "Manutenção programada" if t == "Preventiva" else "Falha inesperada"
    for t in OrderType
]
```

```python
# Se a OS está aberta → Data_Fechamento fica vazia
df.loc[df["Status"] == "Aberta", "Data_Fechamento"] = pd.NaT
```

```python
# Idade da OS
AgeDays = [(f - c).days for c, f in zip(CreationDate, FinishDate)]
```

```python
# STATUS DA ORDEM DE SERVIÇO
# ------------------------

# 30% abertas, 70% fechadas
df["Status"] = np.random.choice(
    ["Aberta", "Fechada"],
    size=len(df),
    p=[0.30, 0.70]
)
```

```python
# -------------------------
# Montagem do DataFrame
# -------------------------

df = pd.DataFrame({
    "OrderID": OrderID,
    "EquipmentID": EquipmentID,
    "EquipmentType": EquipmentType,
    "OrderType": OrderType,
    "Priority": Priority,
    "CreationDate": CreationDate,
    "FinishDate": FinishDate,
    "Status": Status,
    "DowntimeHours": DowntimeHours,
    "EstimatedCost": EstimatedCost,
    "ActualCost": ActualCost,
    "MaintenanceGroup": MaintenanceGroup,
    "Description": Description,
    "AgeDays": AgeDays,
    "Status": Status
})

df.head(20)
```

```python
#Idade media das OS's
df["AgeDays"].mean()
```

```python
#Idade média por tipo de Ordem
df.groupby("OrderType")["AgeDays"].mean()
```

```python
##Idade média por equipamento
df.groupby("EquipmentID")["AgeDays"].mean().sort_values(ascending=False)
```

```python
#exibir os em aberto
df_abertas = df[df["Status"] == "Em execução"]
df_abertas.head()
```

```python
df_abertas_criticas = df[
    (df["Status"] == "Aberta") &
    (df["OrderType"] == "Corretiva")
]

df_abertas_criticas.head()
```

```python
#importando dados para CSV
df.to_csv("ordens_servico.csv", index=False, encoding="utf-8")
```
