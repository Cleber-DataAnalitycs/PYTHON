```python
#Idade média dos equipamentos
# 1) Garantir datetime
df['data_aquisicao_y'] = pd.to_datetime(df['data_aquisicao_y'], errors='coerce', dayfirst=True)

# 2) Hoje (naïve, sem timezone)
hoje = pd.Timestamp.today().normalize()

# 3) Idade em anos
df['idade_anos'] = (hoje - df['data_aquisicao_y']).dt.days / 365.25

# 4) Idade média (ignora NaT automaticamente)
idade_media_anos = df['idade_anos'].mean()
print(f'Idade média dos equipamentos: {idade_media_anos:.2f} anos')
```

```python
#idade media das Ordens de Serviço
# Garantir que data_abertura é datetime
df['data_abertura'] = pd.to_datetime(df['data_abertura'], errors='coerce', dayfirst=True)

# Hoje (sem timezone)
hoje = pd.Timestamp.today().normalize()

# Idade da OS em dias
df['idade_os_dias'] = (hoje - df['data_abertura']).dt.days

# Idade média (ignora NaT automaticamente)
idade_media_dias = df['idade_os_dias'].mean()
print(f'Idade média das OS: {idade_media_dias:.2f} dias')

# Opcional: preparar para gráfico (uma métrica única)
pd.DataFrame({'metric': ['idade_media_os_dias'], 'value': [idade_media_dias]}).to_csv('idade_media_os.csv', index=False)

# Se quiser ver por OS individual (work_order_id)
df[['work_order_id','idade_os_dias']].to_csv('idade_por_os.csv', index=False)
```

```python
# 1) Garantir datas como datetime
df['data_abertura'] = pd.to_datetime(df['data_abertura'], errors='coerce', dayfirst=True)
df['data_fim']      = pd.to_datetime(df['data_fim'],      errors='coerce', dayfirst=True)

# 2) Extrair mês (timestamp do 1º dia do mês)
df['mes_abertura'] = df['data_abertura'].dt.to_period('M').dt.to_timestamp()
df['mes_fim']      = df['data_fim'].dt.to_period('M').dt.to_timestamp()

# 3) Contagens mensais
abertas = (df.groupby('mes_abertura', dropna=True)['work_order_id']
             .size().rename('os_abertas'))
encerradas = (df.loc[df['data_fim'].notna()]
                .groupby('mes_fim')['work_order_id']
                .size().rename('os_encerradas'))

# 4) Juntar e garantir todos os meses no intervalo
inicio = min(abertas.index.min(), encerradas.index.min() if not encerradas.empty else abertas.index.min())
fim    = max(abertas.index.max(), encerradas.index.max() if not encerradas.empty else abertas.index.max())
idx_meses = pd.date_range(inicio, fim, freq='MS')

mensal = (pd.concat([abertas, encerradas], axis=1)
            .reindex(idx_meses)
            .fillna(0)
            .astype(int)
            .rename_axis('mes')
            .reset_index())

print(mensal.head())
# Opcional: salvar
# mensal.to_csv('os_abertas_encerradas_mensal.csv', index=False)
```

```python

# Evita divisão por zero
den = mensal['os_abertas'].replace(0, pd.NA)

mensal['taxa_encerramento'] = (mensal['os_encerradas'] / den).fillna(0).clip(lower=0)
mensal['taxa_encerramento_%'] = (mensal['taxa_encerramento'] * 100).round(2)

taxa_global = mensal['os_encerradas'].sum() / max(mensal['os_abertas'].sum(), 1)

print(mensal[['mes','os_abertas','os_encerradas','taxa_encerramento_%']].head())
print(f"Taxa de encerramento (global): {taxa_global:.2%}")
# mensal.to_csv('os_taxa_encerramento_mensal.csv', index=False)
```
